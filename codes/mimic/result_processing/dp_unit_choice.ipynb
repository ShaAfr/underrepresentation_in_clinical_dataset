```python
train = ['1 unit', '2 unit', '3 unit', '4 unit', '5 unit', '6 unit', '7 unit', '8 unit', '9 unit', '10 unit', '12 unit', '14 unit', '16 unit', '18 unit', '20 unit']

xyz = ['unit_1.csv', 'unit_2.csv', 'unit_3.csv', 'unit_4.csv', 'unit_5.csv', 'unit_6.csv', 'unit_7.csv','unit_8.csv', 'unit_9.csv', 'unit_10.csv', 'unit_12.csv', 'unit_14.csv','unit_16.csv', 'unit_18.csv', 'unit_20.csv']

models = []
for i in xyz:
    demographics = pd.read_csv('temp_' + i)
    models.append(demographics)
```

```python
full_list_C1 = []
full_list_C0 = []
```

```python
#confusion matrix
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import precision_recall_curve
import numpy as np


def result_print_C1(y_true,y_pred,prediction):
  
  recall = recall_score(y_true, y_pred)
  #print("recall_score", recall)
  precision = precision_score(y_true, y_pred)
  #print("precision_score", precision)
  
  k=3
  res = []
  res.append(round(recall,k))
  res.append(round(precision,k))
  
  return res
  
def result_print_C0(y_true,y_pred,prediction):
  
  CM = confusion_matrix(y_true, y_pred)
  TN = CM[0][0]
  FN = CM[1][0]
  TP = CM[1][1]
  FP = CM[0][1]
  FPR = ((FP*0.1)/(FP+TN))
  FNR = ((FN*0.1)/(FN+TP))
  TPR = (TP) / (TP + FN)
  TNR = (TN) / (TN + FP)

  #print("Precision_score (Group 0): ", (TN/(TN+FN)))
  precision_0 = (TN/(TN+FN))
  #print("recall_score (Group 0): ", (TN/(TN+FP)))
  recall_0 = (TN/(TN+FP))
  
  k=3
  
  res = []
  res.append(round(recall_0,k))
  res.append(round(precision_0,k))
  
  
  return res
```

```python
i = 0
for demographics in models:
    res_temp = []
    res_temp.append(train[i])
    
    demographics_age_30_40 = demographics[(demographics['AGE'] >= 30) & (demographics['AGE'] < 40) ]
    
    abc = result_print_C1(demographics_age_30_40["y_true"].values, demographics_age_30_40["y_pred"].values, demographics_age_30_40["prediction"].values)
    res_temp = res_temp + abc
    
    y_pred = demographics["y_pred"].values
    y_true = demographics["y_true"].values
    prediction = demographics["prediction"].values
    #print("For whole group:")
    abc = result_print_C1(y_true,y_pred, prediction)
    res_temp = res_temp + abc
    
    full_list_C1.append(res_temp)
    print(res_temp)
    i=i+1
```

```python
def result_print_performance(y_true,y_pred,prediction):
    CM = confusion_matrix(y_true, y_pred)
    TN = CM[0][0]
    FN = CM[1][0]
    TP = CM[1][1]
    FP = CM[0][1]
    FPR = ((FP*0.1)/(FP+TN))
    FNR = ((FN*0.1)/(FN+TP))
    TPR = (TP) / (TP + FN)
    TNR = (TN) / (TN + FP)
    balanced_accuracy = (TPR + TNR) / 2 

    (precisions, recalls, thresholds) = metrics.precision_recall_curve(y_true, prediction)
    auprc = metrics.auc(recalls, precisions)
    k=3
    
    res_per = []
    res_per.append(round(balanced_accuracy,k))
    res_per.append(round(auprc,k))
    
    return res_per
```

```python
full_list_performance = []
i = 0
for demographics in models:
    res_temp = []
    res_temp.append(train[i])
    demographics_age_30_40 = demographics[(demographics['AGE'] >= 30) & (demographics['AGE'] < 40) ]
    abc = result_print_performance(demographics_age_30_40["y_true"].values, demographics_age_30_40["y_pred"].values, demographics_age_30_40["prediction"].values)
    res_temp = res_temp + abc
    
    y_pred = demographics["y_pred"].values
    y_true = demographics["y_true"].values
    prediction = demographics["prediction"].values
    abc = result_print_performance(y_true,y_pred, prediction)
    res_temp = res_temp + abc
    
    full_list_performance.append(res_temp)
    i=i+1
    
```
