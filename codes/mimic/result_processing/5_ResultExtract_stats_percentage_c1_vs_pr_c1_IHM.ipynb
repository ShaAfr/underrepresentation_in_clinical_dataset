```python
df = pd.read_csv(r"csv_files\original_statistics\temp_train_listfile.csv")
df
```

```python
def different_demographics(df_temp):
    total = df_temp.shape[0]
    positive = df_temp[df_temp['y_true'] == 1].shape[0]
    print(total)
    print(positive)
    print(positive/total)
    return positive/total
    
```

```python
full_list = []

# whole
percentage = different_demographics(df)
a = ['whole']+ [percentage]
full_list.append(a)

# Male
percentage = different_demographics(df[df['GENDER'] == 'M'])
a = ['Male']+ [percentage]
full_list.append(a)

# Female
percentage = different_demographics(df[df['GENDER'] == 'F'])
a = ['Female']+ [percentage]
full_list.append(a)

# White
percentage = different_demographics(df[df['ETHNICITY'].str.contains("WHITE")])
a = ['White']+ [percentage]
full_list.append(a)

# Black
percentage = different_demographics(df[df['ETHNICITY'].str.contains("BLACK")])
a = ['Black']+ [percentage]
full_list.append(a)

# Hispanic
percentage = different_demographics(df[df['ETHNICITY'].str.contains("HISPANIC")])
a = ['Hispanic']+ [percentage]
full_list.append(a)

# Asian
percentage = different_demographics(df[df['ETHNICITY'].str.contains("ASIAN")])
a = ['Asian']+ [percentage]
full_list.append(a)

# age < 30
percentage = different_demographics(df[df['AGE'] < 30])
a = ['Age < 30']+ [percentage]
full_list.append(a)

# 30 >= age > 40
percentage = different_demographics(df[(df['AGE'] >= 30) & (df['AGE'] < 40) ])
a = ['Age [30, 40)']+ [percentage]
full_list.append(a)

# 40 >= age > 50
percentage = different_demographics(df[(df['AGE'] >= 40) & (df['AGE'] < 50) ])
a = ['Age [40, 50)']+ [percentage]
full_list.append(a)

# 50 >= age > 60
percentage = different_demographics(df[(df['AGE'] >= 50) & (df['AGE'] < 60) ])
a = ['Age [50, 60)']+ [percentage]
full_list.append(a)

# 60 >= age > 70
percentage = different_demographics(df[(df['AGE'] >= 60) & (df['AGE'] < 70) ])
a = ['Age [60, 70)']+ [percentage]
full_list.append(a)

# 70 >= age > 80
percentage = different_demographics(df[(df['AGE'] >= 70) & (df['AGE'] < 80) ])
a = ['Age [70, 80)']+ [percentage]
full_list.append(a)


# 80 >= age > 90
percentage = different_demographics(df[(df['AGE'] >= 80) & (df['AGE'] < 90) ])
a = ['Age [80, 90)']+ [percentage]
full_list.append(a)

# age >= 90
percentage = different_demographics(df[df['AGE'] >= 90])
a = ['Age >= 90']+ [percentage]
full_list.append(a)

df_full_list = pd.DataFrame(full_list, columns = ["Group", "percentage_C1"])
print(df_full_list)

# df_full_list.to_csv(model_path + 'result_'+ f, index=False)
    
# print('Metrics csv are posted')
```

```python
df_full_list["Recall_C1"] = Recall_C1
df_full_list["Recall_C0"] = Recall_C0
df_full_list["Balanced_Accuracy"] = Balanced_Accuracy
df_full_list["AUC_ROC"] = AUC_ROC
```
