```python
from os import listdir
from os.path import isfile, join
import fnmatch
import pandas as pd

# *************** change dp name in the following line ******************** 
models = ['original', 'dp_black_avg', 'gamma_2',  'adasyn', 'smote', 'oversampling', 'undersampling', 'nearmiss1', 'nearmiss3', 'distant' ]

group = 4 # black=4, age>=90=14
```

```python
full_list = []
for mod in models:
    
    model_path = r"csv_files/"+ mod + "/test/"

    
    files = [f for f in listdir(model_path) if fnmatch.fnmatch(f, '*result*.csv')]

    # take the csv files in that directory
#     if 'dp' in mod:
#         files = [f for f in listdir(model_path) if fnmatch.fnmatch(f, '*result_temp_test_isotonic*.csv')]
#     else:
#         files = [f for f in listdir(model_path) if fnmatch.fnmatch(f, '*AVERAGE*.csv')]
    
    for file in files:
        temp_list = []
        df_individual = pd.read_csv(model_path  + file)
        temp_list.append(mod)
        temp_list.append(df_individual.iloc[group,2])  # recall_C1
        temp_list.append(df_individual.iloc[group,3])  # Precision_C1
        temp_list.append(df_individual.iloc[group,5])  # F1_C1
        temp_list.append(df_individual.iloc[group,4]) # PR_C1
        temp_list.append(df_individual.iloc[group,11])  # Balanced_accuracy_C1
        temp_list.append(df_individual.iloc[group,13])  # MCC
        full_list.append(temp_list)
    
    print("Group name:", df_individual.iloc[group,0])
    
    

df_full = pd.DataFrame(full_list, columns= ["model", "Recall_C1", "Prec_C1", "F1_C1","PR_C1", "bal_acc", "MCC"])
```
