# 1. Disparity (group-wise): highest and lowest  (For HeatMap figure)

```python
from os import listdir
from os.path import isfile, join
import fnmatch
import pandas as pd


# *************** change dp name in the following line ******************** 
models = ['original', 'gamma_2', 'adasyn', 'smote', 'oversampling',  'undersampling', 'nearmiss1', 'nearmiss3', 'distant']


metric_idx = 13 #F1_C1 = 5 ; Rec_C1 = 2 ; bal_acc = 11; MCC = 13
evaluation = "relative_disparity" # "relative_disparity" or "absolute_disparity"
```

```python
full_list = []
for mod in models:
    model_path = r"csv_files/"+ mod + "/test/"



    # take the csv files in that directory
    if 'dp' in mod:
        files = [f for f in listdir(model_path) if fnmatch.fnmatch(f, '*result_temp_test_isotonic*.csv')]
    else:
        files = [f for f in listdir(model_path) if fnmatch.fnmatch(f, '*AVERAGE*.csv')]
    print(files)
    
    df = pd.read_csv(model_path  + files[0])
    
    #full_list.append(list(df.loc[4:6]))
    values = df.iloc[7:,metric_idx]  #3:7 -- ethnicity ; 7: --> age ; 1:3 --> gender
    print(values)
    if evaluation == "relative_disparity":
        disparity = max(list(values)) / min(list(values))
    elif evaluation== "absolute_disparity":
        disparity = max(list(values)) - min(list(values))
    else:
        print("Error")
        break
        
    
    print(disparity)
    full_list.append(disparity)
    

df_full = pd.DataFrame(full_list)
```

```python
# printing one row of result 
# For example, the absoute disparity results from all 9 sampling (showing in row). 
# (0,0) represents original model disparity among one group (age, gender, race). If asian has highest F1 score and black has lowest F1 score, then absolute disparity is (Asian F1 - Black F1) 
df_full
```

```python
from os import listdir
from os.path import isfile, join
import fnmatch
import pandas as pd


# *************** change dp name in the following line ******************** 
models = ['dp_age30_avg', 'dp_black_avg', 'dp_asian_avg', 'dp_age4050_avg', 'dp_age90_avg', 'dp_hispanic_avg']
indices = [7,4,6,9,14,5]
Groups = ['Age<30', 'Black', 'Asian', 'Age [40, 50)', 'Age>=90', 'Hispanic']


metric_idx = 2 #F1_C1 = 5 ; Rec_C1 = 2 ; bal_acc = 11; MCC = 13


```

```python
full_list = []
i=0
for mod in models:
    model_path = r"csv_files/"+ mod + "/test/"

    # take the csv files in that directory
    
    files = [f for f in listdir(model_path) if fnmatch.fnmatch(f, '*result_temp_*.csv')]
    
    for file in files:
        
        df = pd.read_csv(model_path  + file)
        values = df.iloc[indices[i],metric_idx]
    
        full_list.append([Groups[i], values])
    
    i=i+1
    

df_full = pd.DataFrame(full_list, columns= ["Group", "DP"])
print(df_full)
```

```python
df_full.to_csv('fig_7_recall_comparison_DP.csv')
```
