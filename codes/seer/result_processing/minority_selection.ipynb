```python
def count_training_percent(cancer):
    minor = 0 if (cancer == 'breast') else 1

    X_train = np.load(sys_path + '/' + cancer + '/X_train.npy')
    y_train = np.load(sys_path + '/' + cancer + '/y_train.npy')
    print('total: ', len(X_train), len(y_train))

    ind1 = np.where(y_train == 1) 
    y_train_1 = y_train[ind1]
    print('class 1: ', len(y_train_1))

    f = open(sys_path + '/' + cancer + '/input_columns.txt', 'r')
    cols = f.readlines()

    gender_index = cols.index('Sex 1\n')
    ind_m = np.where(X_train[:,gender_index] == 1)
    y_train_m = y_train[ind_m]
    print('male: ', len(y_train_m))
    ind = np.where(y_train_m == minor)
    y_train_m_minor = y_train_m[ind]
    print(len(y_train_m_minor), len(y_train_m_minor) / len(y_train_m))
    ind_f = np.where(X_train[:,gender_index] == 0)
    y_train_f = y_train[ind_f]
    print('female: ', len(y_train_f))
    ind = np.where(y_train_f == minor)
    y_train_f_minor = y_train_f[ind]
    print(len(y_train_f_minor), len(y_train_f_minor) / len(y_train_f))

    white_index = cols.index('Race recode Y 1\n')
    ind_w = np.where(X_train[:,white_index] == 1)
    y_train_w = y_train[ind_w]
    print('wihte: ', len(y_train_w))
    ind = np.where(y_train_w == minor)
    y_train_w_minor = y_train_w[ind]
    print(len(y_train_w_minor), len(y_train_w_minor) / len(y_train_w))

    black_index = cols.index('Race recode Y 2\n')
    ind_b = np.where(X_train[:,black_index] == 1)
    y_train_b = y_train[ind_b]
    print('black: ', len(y_train_b))
    ind = np.where(y_train_b == minor)
    y_train_b_minor = y_train_b[ind]
    print(len(y_train_b_minor), len(y_train_b_minor) / len(y_train_b))

    hisp_index = cols.index('Origin Recode NHIA 1\n')
    ind_h = np.where(X_train[:,hisp_index] == 1)
    y_train_h = y_train[ind_h]
    print('hispanic: ', len(y_train_h))
    ind = np.where(y_train_h == minor)
    y_train_h_minor = y_train_h[ind]
    print(len(y_train_h_minor), len(y_train_h_minor) / len(y_train_h))

    asian_index = cols.index('Race recode Y 4\n')
    ind_a = np.where(X_train[:,asian_index] == 1)
    y_train_a = y_train[ind_a]
    print('asian: ', len(y_train_a))
    ind = np.where(y_train_a == minor)
    y_train_a_minor = y_train_a[ind]
    print(len(y_train_a_minor), len(y_train_a_minor) / len(y_train_a))

    other_ind = np.where((X_train[:,white_index] == 0) & (X_train[:,asian_index] == 0) & (X_train[:,black_index] == 0) & (X_train[:,hisp_index] == 0))
    other = y_train[other_ind]
    print('other: ', len(other))

    age = X_train[:,0]
    ind_30 = np.where(age < 30)
    y_train_30 = y_train[ind_30]
    print('<30: ', len(y_train_30))
    ind = np.where(y_train_30 == minor)
    y_train_30_minor = y_train_30[ind]
    print(len(y_train_30_minor), len(y_train_30_minor) / len(y_train_30))

    ind_30_40 = np.where((age >= 30) & (age < 40))
    y_train_30_40 = y_train[ind_30_40]
    print('30-40: ', len(y_train_30_40))
    ind = np.where(y_train_30_40 == minor)
    y_train_30_40_minor = y_train_30_40[ind]
    print(len(y_train_30_40_minor), len(y_train_30_40_minor) / len(y_train_30_40))

    ind_40_50 = np.where((age >= 40) & (age < 50))
    y_train_40_50 = y_train[ind_40_50]
    print('40-50: ', len(y_train_40_50))
    ind = np.where(y_train_40_50 == minor)
    y_train_40_50_minor = y_train_40_50[ind]
    print(len(y_train_40_50_minor), len(y_train_40_50_minor) / len(y_train_40_50))

    ind_50_60 = np.where((age >= 50) & (age < 60))
    y_train_50_60 = y_train[ind_50_60]
    print('50-60: ', len(y_train_50_60))
    ind = np.where(y_train_50_60 == minor)
    y_train_50_60_minor = y_train_50_60[ind]
    print(len(y_train_50_60_minor), len(y_train_50_60_minor) / len(y_train_50_60))

    ind_60_70 = np.where((age >= 60) & (age < 70))
    y_train_60_70 = y_train[ind_60_70]
    print('60-70: ', len(y_train_60_70))
    ind = np.where(y_train_60_70 == minor)
    y_train_60_70_minor = y_train_60_70[ind]
    print(len(y_train_60_70_minor), len(y_train_60_70_minor) / len(y_train_60_70))

    ind_70_80 = np.where((age >= 70) & (age < 80))
    y_train_70_80 = y_train[ind_70_80]
    print('70-80: ', len(y_train_70_80))
    ind = np.where(y_train_70_80 == minor)
    y_train_70_80_minor = y_train_70_80[ind]
    print(len(y_train_70_80_minor), len(y_train_70_80_minor) / len(y_train_70_80))

    ind_80_90 = np.where((age >= 80) & (age < 90))
    y_train_80_90 = y_train[ind_80_90]
    print('80-90: ', len(y_train_80_90))
    ind = np.where(y_train_80_90 == minor)
    y_train_80_90_minor = y_train_80_90[ind]
    print(len(y_train_80_90_minor), len(y_train_80_90_minor) / len(y_train_80_90))

    ind_90 = np.where(age >= 90)
    y_train_90 = y_train[ind_90]
    print('>=90: ', len(y_train_90))
    ind = np.where(y_train_90 == minor)
    y_train_90_minor = y_train_90[ind]
    print(len(y_train_90_minor), len(y_train_90_minor) / len(y_train_90))
```
